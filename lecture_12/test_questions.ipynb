### Question 1
What is the value of count after execution of the following code?

```python
data = [4,4,5,5,6,6,6,7,7,7,7,8,8]
count = 0
for i in range(1, len(data)):
    if data[i] == data[i-1]:
        count = count + 1

print(count)
```

### Question 3
What will this program return for the following values? rek1(3, 4), rek1(3, -4) and rek1(-3, 4)

```python
def rek1(s, t):
    if s > 0:
        r = rek1(s - 1, t) + t
    else:
        r = 0
    return r

print(rek1(3, 4))
print(rek1(3, -4))
print(rek1(-3, 4))
```

### Question 4
What is the value of p after the following code runs?

```python
count = 6
x = [1,3,0,2,2,4]
p = 0
for i in range(count):
    if x[i] > x[p]:
        p += 1

print(p)
```

### Question 6
Which statement is true for the following code?
- if x > 0 and y > x function f will never stop
- if y <= 0 function f will never stop
- if x < 0 function f will never stop
- if x > 0 and y > 0 function f will never stop
- if x > 0 and y < x function f will never stop

```python
def f(x, y):
    while x > 0:
        x -= 2 * y

f(1,2)
f(1, 0)
f(-1, 2)
f(1,1)
f(2, 1)
```
