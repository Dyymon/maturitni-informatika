# Objects

Without us knowing we have been using objects, or classes for quite some time, for example, many python structures like dictionary and even lists behave like objects.

Now what are objects? They are one of the new building blocks we can use in our coding. These entities not only store values but functions as well.

```python
dictionary = {"a": 1, "b": 2, "c": 3}
print(dictionary.keys())
print(dictionary.pop("a"))
```

As we can see dictionary stores things but has some functions as well
