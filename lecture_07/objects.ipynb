# Objects

Without us knowing we have been using objects, or classes for quite some time, for example, many python structures like dictionary and even lists behave like objects.

Now what are objects? They are one of the new building blocks we can use in our coding. These entities not only store values but functions as well.

```python
dictionary = {"a": 1, "b": 2, "c": 3}
print(dictionary.keys())
print(dictionary.pop("a"))
```

As we can see dictionary stores things but has some functions as well

## Object Student

let us make our own object called student, this object is initiated via __init()__ function, meaning that by calling __Student(args**)__ we are actually calling the __init()__ function.

Another important part are the __self__ arguments, __self__ specifically refers to the __class__ the object it is in. Meaning that if we pass __self__ as an argument into a function, we can use other functions or variables of the __class__ inside it.

If the function of the class doesn't get a __self__ argument, it is considered static and the __@staticmethod__ argument needs to be written abow it.

If we want a function or argument not to be automatically visible for other users when writing and we want to show them that this function or argument is mainly for inside workings of the __class__ we use underscore to mark it as so as we can see with the ___easter_egg()__ function or ___grades__ argument.

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.grade = 1
        self._grades = []

    def add_grade(self, grade):
        self._grades.append(grade)
        self.grade = sum(self._grades) / len(self._grades)

    @staticmethod
    def _easter_egg():
        print("pepa bude v televizi")

pepa = Student("Pepa", 27)
print("Pepas grade is: ", pepa.grade)
pepa.add_grade(1.)
pepa.add_grade(2.)
pepa.add_grade(5.)
print("pepas grade after grading is: ", pepa.grade)
print("these are pepas grades: ", pepa._grades)
pepa._easter_egg()

```

## More examples

Objects in python have some predefined functions that do a specific thing. We already know __init__ which needs to be present for the __class__ to even work.
You can define functions like __mul__ or __add__ to define what happens if you invoke operations such as __+__ or __*__ with the class. Example bellow.

```python
class Number:

    def __init__(self, number):
        self.number = number

    def __mul__(self, other):
        return self.number + other.number

    def __add__(self, other):
        return self.number * other.number

a = Number(2)
b = Number(3)

print(a * b)
print(a + b)
```
