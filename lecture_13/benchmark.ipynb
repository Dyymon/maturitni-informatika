# Benchmark sorting Algorithms
Down here we have a script that benchmarks student solutions against each other, against simple bubble sort and against python native sort

```python

import random
import time
import matplotlib.pyplot as plt

from matej_s import sort_numbers as matej
from veronika_h import sort_numbers as veronika
from lukas_h import sort_numbers as lukas
from filip_c import sort_numbers as filip
from eva_v import sort_numbers as eva
from bubble import sort_numbers as bubble


# ---- CONFIG ----
SIZES = [100, 500, 1000, 2000, 5000, 10000]
REPEATS = 5
MAX_INT = 100000

def pysort(numbers):
    return sorted(numbers)


def benchmark(func, sizes):
    times = []

    for n in sizes:
        total_time = 0

        for _ in range(REPEATS):
            data = [random.randint(0, MAX_INT) for _ in range(n)]
            start = time.perf_counter()
            func(data.copy())
            end = time.perf_counter()

            total_time += (end - start)

        avg_time = total_time / REPEATS
        times.append(avg_time)

    return times


def main():
    results = {
        "Matej": benchmark(matej, SIZES),
        "Veronika": benchmark(veronika, SIZES),
        "Lukas": benchmark(lukas, SIZES),
        "Filip": benchmark(filip, SIZES),
        "Eva": benchmark(eva, SIZES),
        "Bubble": benchmark(bubble, SIZES),
        "Python built-in": benchmark(pysort, SIZES),
    }

    # ---- Plot 1: Full scale ----
    plt.figure()
    for name, times in results.items():
        plt.plot(SIZES, times, label=name)

    plt.xlabel("Input size (n)")
    plt.ylabel("Time (seconds)")
    plt.title("Sorting Benchmark (Full Scale)")
    plt.legend()
    plt.grid(True)
    plt.show()

    # ---- Plot 2: Zoomed (0–0.1 sec) ----
    plt.figure()
    for name, times in results.items():
        plt.plot(SIZES, times, label=name)

    plt.xlabel("Input size (n)")
    plt.ylabel("Time (seconds)")
    plt.title("Sorting Benchmark (Zoomed to 0–0.1 s)")
    plt.ylim(0, 0.1)
    plt.legend()
    plt.grid(True)
    plt.show()

    # ---- Plot 2: Zoomed (0–0.1 sec) ----
    plt.figure()
    for name, times in results.items():
        plt.plot(SIZES, times, label=name)

    plt.xlabel("Input size (n)")
    plt.ylabel("Time (seconds)")
    plt.title("Sorting Benchmark (Zoomed to 0–0.02 s)")
    plt.ylim(0, 0.02)
    plt.legend()
    plt.grid(True)
    plt.show()




if __name__ == "__main__":
    main()

```
