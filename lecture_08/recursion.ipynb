# Recursion
Not only you can pass a function as an argument or return functions result as an argument, you can return the result of the function you are defining by calling it again.

Recursion is: Recursion is: Recursion is: Recursion is: Recursion is: This is recursion.

Let's go over few showcases of recursion, the first, adding numbers until we reach a certain amount.

```python
def adding_recursion(start, limit):
    if start == limit:
        return start
    return adding_recursion(start+1, limit)

adding_recursion(10, 100)
```

Or just repeating until we reach depth zero

```python
def recursion(depth):
    if depth == 0:
        return "done"
    return recursion(depth - 1)

recursion(100)
```

## Factorial
You can find a simple albeit useful case for a recursion in counting factorial of a certain number.

```python
def factorial(num):
    if num == 0:
        return 1
    return num * factorial(num-1)

factorial(3)
```

## Path search in tree

Here we have an example of how to use recursion for path planning, this is an example of Depth First Search (DFS).

```python
tree = {0: [1, 2], 1: [3, 4 , 5], 2: [6], 3: [7, 8], 4: [], 5: [9], 6: [10, 11, 12, 13], 7: [], 8: [14, 15], 9: [], 10: [], 11: [16, 17], 12: [], 13: [18], 14: [], 15: [], 16: [19], 17: [], 18: [20], 19: [], 20: []}

def find_path(goal_id, path, cur_id=0):
    path.append(cur_id)
    if cur_id == goal_id:
        return path

    children = tree[cur_id]
    if len(children) == 0:
        path.pop(-1)
        return None

    for child in children:
        new_path = find_path(goal_id, path, child)
        if new_path is not None:
            return new_path

    path.pop(-1)
    return None


goal = 20
start = 0
find_path(goal, [], cur_id=start)


```
